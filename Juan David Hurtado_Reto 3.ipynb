```python
###-----------------------------------------------------------------------------
#Punto 1 Reto No3
###-----------------------------------------------------------------------------
# Creación de datos de cultivos
###-----------------------------------------------------------------------------
nombre_cultivo = ['Papa','Zanahoria','Yuca']
dia_hora_mante = ['Lunes, Jueves 4:00 pm', 'Martes, Miercoles 2:00 pm', 
                  'Miercoles, Viernes 10:00 am'
                  ]
dia_hora_regad = ['Martes, Viernes 1:00 pm', 'Martes, Jueves 5:00 pm', 
                  'Lunes, Jueves 9:00 am'
                  ]
dia_hora_abono = ['Miercoles, Sabado 11:00 am','Domingo, Martes 1:00 pm', 
                  'Martes, Sabado 11:00 am'
                  ]
etapa_1 = ['Etapa 1 – Fase Vegetativa, 5 a 30 días', 'Etapa 1 – Fase Vegetativa, 10 a 20 días', 
           'Etapa 1 – Fase Vegetativa, 20 a 30 días'
           ]
etapa_2 = ['Etapa 2 - Reproductiva, 10 a 40 dias', 'Etapa 2 - Reproductiva, 5 a 20 dias', 
           'Etapa 2 - Reproductiva, 25 a 35 dias'
           ]
etapa_3 = ['Etapa 3 - Maduración, 20 a 30 dias', 'Etapa 3 - Maduración, 5 a 10 dias', 
           'Etapa 3 - Maduración, 10 a 20 dias'
           ]

###-----------------------------------------------------------------------------
# Declaración de variables Punto 1 y Punto 2
###-----------------------------------------------------------------------------

#Variables Gastos (Costos Variables)
lista_gastos = []
medicamentos = []
imprevistos = []
#Variables Costos (Costos Fijos)
lista_costos = []
mano_obra = []
abono = []
agua = []
mantenimiento = []
#Variables producción
valor_arroba = []
cantidad_kilos =[]
#inicialización variables cálculos
costo_total = 0
costos_totales = 0
gasto_total = 0
meses_sin_gasto = []
meses_menos_100000 = []

###-----------------------------------------------------------------------------
### Función para ingresar datos del cultivo
###-----------------------------------------------------------------------------

def ingresar_datos(lista_datos,nombre_lista_datos):
    ''' Función para ingresar los datos
        entrada 1: lista de datos
        entrada 2: nombre lista datos
    '''
    for opcion_3 in range(5):
      dato = int(input(f'Ingresar el valor de {nombre_lista_datos} del mes {opcion_3+1}: '))
      lista_datos.append(dato)
    print(f'Los datos ingresados en {nombre_lista_datos} son {lista_datos}\n') 

###-----------------------------------------------------------------------------
#Ciclo While para el menú de opciones
###-----------------------------------------------------------------------------

while True:
  opcion=0
  opcion = int(input(print(f'Por favor ingrese la opción que desea:\n'
                            '1. Horario de Gestión de Cultivo\n'
                            '2. Etapas del cultivo\n'
                            '3. Información contable\n'
                            '4. Salir'
              )))
  ###--------------------------------------------------------------------------- 
  #Opción 1, Imprimir Horario de Gestión de Cultivo
  ###---------------------------------------------------------------------------   
  if opcion==1:
    print('Seleccione un cultivo: ')
    for i, nombres_cultivos in enumerate(nombre_cultivo):
      print(f'{i+1}.{nombres_cultivos}')
    opcion_1 = int(input())-1
    print(f'Dias y hora de mantenimiento: {dia_hora_mante[opcion_1]}')
    print(f'Dias y hora de regado: {dia_hora_regad[opcion_1]}')
    print(f'Dias y hora de abono: {dia_hora_abono[opcion_1]}\n')
  ###---------------------------------------------------------------------------
  #Opción 2, Imprimir Etapas del cultivo
  ###---------------------------------------------------------------------------   
  if opcion==2:
    print('Seleccione un cultivo: ')
    for i, nombres_cultivos in enumerate(nombre_cultivo):
      print(f'{i+1}.{nombres_cultivos}')
    opcion_2 = int(input())-1
    print(f'{etapa_1[opcion_2]}')
    print(f'{etapa_2[opcion_2]}')
    print(f'{etapa_3[opcion_2]}\n')
  
  ###---------------------------------------------------------------------------
  #Opción 3
  ###---------------------------------------------------------------------------
  ###PUNTO 2 Reto No3
  ###---------------------------------------------------------------------------
  if opcion==3:
    print('Ingresar gastos: \n')
    ingresar_datos(medicamentos,'medicamentos')
    ingresar_datos(imprevistos,'imprevistos')
    print('Ingresar costos: \n')
    ingresar_datos(mano_obra,'mano de obra')
    ingresar_datos(abono,'abono')
    ingresar_datos(agua,'agua')
    print('Ingresar producción: \n')
    ingresar_datos(valor_arroba,'Valor Arroba')
    ingresar_datos(cantidad_kilos,'Cantidad de kilos')
    
    ###-------------------------------------------------------------------------
    #Costos totales del cultivo por mes
    ###-------------------------------------------------------------------------
    for mes,costo in enumerate(zip(medicamentos,imprevistos,abono,mano_obra,agua)):
        costo_total = sum(costo) 
        lista_costos.append(costo_total)
        print(f'El costo total del mes {mes+1} es: {costo_total} ')
    print(f'Los costos totales son: {lista_costos}')

    ###-------------------------------------------------------------------------
    #Costos totales de mano de obra
    ###-------------------------------------------------------------------------
    costos_mano_obra = sum(mano_obra)

    ###-------------------------------------------------------------------------
    #Meses sin gastos
    ###-------------------------------------------------------------------------
    for mes,gasto in enumerate(zip(medicamentos,imprevistos)):
        gasto_total = sum(gasto)
        lista_gastos.append(gasto_total)
        if gasto_total == 0:
          meses_sin_gasto.append(mes+1)
        if gasto_total < 100000:
          meses_menos_100000.append(mes+1)
    print(f'Los gastos totales son: {lista_gastos}') 
    print(f'Los meses sin gastos fueron:')
    for mes in meses_sin_gasto:
      print(f'mes {mes}')
    
    ###-------------------------------------------------------------------------
    #Meses gastos menores a 100.000
    ###-------------------------------------------------------------------------
    print('Los meses con gastos menores a 100.000 fueron:')
    for mes in meses_menos_100000:
      print(f'mes {mes}')

    ###-------------------------------------------------------------------------  
    #Cálculo del valor promedio de costos Fijos y costos variables
    ###-------------------------------------------------------------------------
    prom_costos = sum(lista_costos)/len(lista_costos)
    4
    print(f'El valor promedio de los costos fijos (costos) es {prom_costos}')
    prom_gastos = sum(lista_gastos)/len(lista_gastos)
    print(f'El valor promedio de los costos variabeles (gastos) es {prom_gastos}')
    
    ###-------------------------------------------------------------------------
    #Cálculo de la Ganancia
    ###-------------------------------------------------------------------------
    lista_ingresos = [valor*cantidad*12.5 for valor,cantidad in zip(valor_arroba,cantidad_kilos)]
    sum_ingresos = sum(lista_ingresos)
    sum_costos_gastos = sum(lista_costos) + sum(lista_gastos)
    ganancia = sum_ingresos - sum_costos_gastos
    if ganancia>=0:
      print(f'La ganancia fue de {ganancia}')
    else: 
      print(f'La perdida fue de {-ganancia}')

    ###-------------------------------------------------------------------------
    #Ganancia si arroz incrementa 37%
    ###-------------------------------------------------------------------------
    sum_ingresos_incremento = sum_ingresos*1.37
    ganancia_incremento = sum_ingresos_incremento - sum_costos_gastos
    if ganancia_incremento>=0:
      print(f'La ganancia con un incremento del 37% del arroz sería de {ganancia_incremento}')
    else: 
      print(f'La perdida con un incremento del 37% del arroz sería de {-ganancia_incremento}')
    
    ###-------------------------------------------------------------------------
    #Ganancia si los costos y gastos se disminuyen en un 5% 
    #y la cantidad de arrobas producidas en un 63%
    ###-------------------------------------------------------------------------
    
    ganancia_especial = sum_ingresos*0.95 - sum_costos_gastos*0.63
    if ganancia_especial>=0:
      print(f'La ganancia disminuyendo 5% cosotos y 63% las arrobas del 37% son de {ganancia_especial}\n')
    else: 
      print(f'La perdida disminuyendo 5% cosotos y 63% las arrobas del 37% son de {-ganancia_especial}\n')
  ###---------------------------------------------------------------------------  
  #Opción 4 
  ###---------------------------------------------------------------------------
  #Salir
  ###---------------------------------------------------------------------------
  if opcion==4:
    break    
```
